1\. Create a random list of number and then save it to a text file named "simple_data.txt"

```python
random_list = np.random.random(100)
new_file = "simple_data.txt"
with open(new_file, mode='w') as f:
    for i in range(random_list.size):
        f.write(str(random_list[i])+'\t')
        if (i+1) % 10 == 0:
            f.write('\n')
print("Operation completed!")
```

2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
random_matrix = np.random.randn(5, 5)
new_file = "data.txt"
with open(new_file, mode='w') as f:
    for i in range(random_matrix.shape[0]):
        for j in range(random_matrix.shape[1]):
            f.write(str(random_matrix[i][j])+'\t')
        f.write('\n')
print("Operation completed!")
```

3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
with open(file="data.txt", mode='r') as f:
    with open(file="data.csv", mode='w') as d:
        for line in f:
            d.write(line)
print("Operation completed!")
```
